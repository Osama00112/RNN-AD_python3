```python
# Load the CSV file into a pandas DataFrame
df2 = pd.read_csv('output/prediction_test.csv')

# Define a function to determine the majority class
def get_majority_class(row):
    probs = {'CN': row['CN relative probability'], 'MCI': row['MCI relative probability'], 'AD': row['AD relative probability']}
    return max(probs, key=probs.get)

# Find unique patient IDs (PIDs)
unique_rids = df2['RID'].unique()

# List to store PIDs with the required majority shift
shift_rids = []

# Iterate through each patient
for pid in unique_rids:
    patient_data = df2[df2['RID'] == pid]
    
    # Get majority class for the first and last month
    first_month_majority = get_majority_class(patient_data[patient_data['Forecast Month'] == 1].iloc[0])
    last_month_majority = get_majority_class(patient_data[patient_data['Forecast Month'] == 84].iloc[0])
    
    # Check for the required majority shift
    if (first_month_majority == 'CN' and last_month_majority == 'MCI'):
        shift_rids.append(pid)
        
    # if (first_month_majority == 'MCI' and last_month_majority == 'AD'):
    #     shift_rids.append(pid)

# Print the result
print("Patient IDs with the required majority probability shift:")
print(shift_rids)
```

```python
# from df_sorted, select only the rows with the required majority shift
df_shifted = df_sorted[df_sorted['RID'].isin(shift_rids)]
df_shifted.head()
```

```python
df_shifted.shape
```

```python
# for each patient, sort by VISCODE
df_shifted_sorted = df_shifted.sort_values(by=['RID', 'VISCODE'])
df_shifted_sorted.head()
```

```python
# for RID 1098, list the RAVLT_immediate values against VISCODE
df_shifted_sorted[df_shifted_sorted['RID'] == 1098][['VISCODE', 'CDRSB', 'RAVLT_immediate']]
def sort_viscode(viscode):
    if viscode == 'bl':
        return (0, 0)  # "bl" should come first
    else:
        num = int(viscode[1:])  # Extract numeric part and convert to integer
        return (1, num)
    
df_shifted['VISCODE_SORT'] = df_shifted['VISCODE'].apply(sort_viscode)
df_shifted_sorted = df_shifted.sort_values(by=['RID', 'VISCODE_SORT'])
df_shifted_sorted.head()

#plot
import matplotlib.pyplot as plt
plt.plot(df_shifted_sorted[df_shifted_sorted['RID'] == 1098]['VISCODE'], df_shifted_sorted[df_shifted_sorted['RID'] == 1098]['RAVLT_immediate'])
plt.xlabel('VISCODE')
plt.ylabel('RAVLT_immediate')
plt.title('RAVLT_immediate values for RID 1098')
plt.show()

```

```python
# for all patients, plot the RAVLT_immdiate score over time in a single plot
import matplotlib.pyplot as plt

# Find unique patient IDs (PIDs)
unique_rids = df_shifted_sorted['RID'].unique()

# Plot the RAVLT_immediate score for each patient
plt.figure(figsize=(10, 6))
for pid in unique_rids:
    patient_data = df_shifted_sorted[df_shifted_sorted['RID'] == pid]
    plt.plot(patient_data['VISCODE'], patient_data['RAVLT_immediate'], label=f'Patient {pid}')
    
    
plt.xlabel('Visit code')
plt.ylabel('RAVLT_immediate score')
plt.title('RAVLT_immediate score over time')
plt.legend()
plt.show()
```
