```python
# Load the CSV file into a pandas DataFrame
df2 = pd.read_csv('output/prediction_test.csv')

# Define a function to determine the majority class
def get_majority_class(row):
    probs = {'CN': row['CN relative probability'], 'MCI': row['MCI relative probability'], 'AD': row['AD relative probability']}
    return max(probs, key=probs.get)

# Find unique patient IDs (PIDs)
unique_rids = df2['RID'].unique()

# List to store PIDs with the required majority shift
shift_rids = []
CN_to_CN = []
MCI_to_MCI = []
AD_to_AD = []
MCI_to_AD = []
CN_to_AD = []
CN_to_MCI = []


for pid in unique_rids:
    patient_data = df2[df2['RID'] == pid]
    
    # Get majority class for the first and last month
    first_month_majority = get_majority_class(patient_data[patient_data['Forecast Month'] == 1].iloc[0])
    last_month_majority = get_majority_class(patient_data[patient_data['Forecast Month'] == 84].iloc[0])
    
    # Check for the required majority shift
    if (first_month_majority == 'CN' and last_month_majority == 'MCI'):
        shift_rids.append(pid)
        CN_to_MCI.append(pid)
        
    
    if (first_month_majority == 'CN' and last_month_majority == 'CN'):
        shift_rids.append(pid)
        CN_to_CN.append(pid)
    
    if (first_month_majority == 'MCI' and last_month_majority == 'MCI'):
        shift_rids.append(pid)
        MCI_to_MCI.append(pid)
        
    if (first_month_majority == 'AD' and last_month_majority == 'AD'):
        shift_rids.append(pid)
        AD_to_AD.append(pid)
        
    if (first_month_majority == 'MCI' and last_month_majority == 'AD'):
        shift_rids.append(pid)
        MCI_to_AD.append(pid)
    
    if (first_month_majority == 'CN' and last_month_majority == 'AD'):
        shift_rids.append(pid)
        CN_to_AD.append(pid)    

# Iterate through each patient
for pid in unique_rids:
    patient_data = df2[df2['RID'] == pid]
    
    # Get majority class for the first and last month
    first_month_majority = get_majority_class(patient_data[patient_data['Forecast Month'] == 1].iloc[0])
    last_month_majority = get_majority_class(patient_data[patient_data['Forecast Month'] == 84].iloc[0])
    
    # Check for the required majority shift
    # if (first_month_majority == 'CN' and last_month_majority == 'MCI'):
    #     shift_rids.append(pid)
    
    # Check for the required majority shift
    # if (first_month_majority == 'CN' and last_month_majority == 'CN'):
    #     shift_rids.append(pid)
    
    # if (first_month_majority == 'MCI' and last_month_majority == 'MCI'):
    #     shift_rids.append(pid)
        
    if (first_month_majority == 'AD' and last_month_majority == 'AD'):
        shift_rids.append(pid)
        
        
    # if (first_month_majority == 'MCI' and last_month_majority == 'AD'):
    #     shift_rids.append(pid)
    
    
    # if (first_month_majority == 'CN' and last_month_majority == 'AD'):
    #     shift_rids.append(pid)

# Print the result
print("Patient IDs with the required majority probability shift:")
print(shift_rids)
```

```python
# convert bl, m12 etc to float for finding trend
def convert_to_float(viscode):
    if viscode == 'bl':
        return 0
    else:
        return float(viscode[1:])
    
df_shifted_sorted['VISCODE_FLOAT'] = df_shifted_sorted['VISCODE'].apply(convert_to_float)
df_shifted_sorted.head()
```

```python

# check if their is a trend in the feature value over time
import numpy as np
from sklearn.linear_model import LinearRegression

# Find unique patient IDs (PIDs)
unique_rids = df_shifted_sorted['RID'].unique()

# List to store the trend for each patient
trend = []

#handle if input y contains NaN
df_shifted_sorted = df_shifted_sorted.dropna(subset=['RAVLT_immediate'])



# Iterate through each patient
for pid in unique_rids:
    patient_data = df_shifted_sorted[df_shifted_sorted['RID'] == pid]
    
    # Fit a linear regression model
    model = LinearRegression()
    model.fit(patient_data[['VISCODE_FLOAT']], patient_data['RAVLT_immediate'])
    
    # Get the trend
    trend.append(model.coef_[0])
    
# Print the result
print("Trend for each patient:")
print(trend)
```
